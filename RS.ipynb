```python
def popularity_rec(data):
    ratings_mean_count = pd.DataFrame(data.groupby('product')['score'].mean())
    ratings_mean_count['rating_counts'] = data.groupby('product')['score'].count()
    ratings_mean_count = ratings_mean_count.sort_values(by=['score','rating_counts'], ascending=[False, False])
    print('Popularity based top 5 recommend mobile phones.: \n')
    display(ratings_mean_count.head())
    return
```

```python
from surprise import Dataset,Reader
from surprise.model_selection import train_test_split
from surprise import SVD, KNNWithMeans
from surprise import accuracy
```

```python
df_final = df_50
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(df_final[['author', 'product', 'score']], reader)
```

```python
trainset, testset = train_test_split(data, test_size=.25, random_state=120)
trainset.all_ratings()
```

```python
print(trainset.to_raw_uid(0))
```

```python
svd_model = SVD(n_factors=5, biased=False)
svd_model.fit(trainset)
```

```python
testset[0]
```

```python
svd_test_pred = svd_model.test(testset)
accuracy.rmse(svd_test_pred)
```

KNN Item based

```python
def knn_item(train, test):
    knn_i = KNNWithMeans(k=10, sim_options={'name': 'pearson_baseline', 'user_based': False})
    knn_i.fit(train)
    knn_i_pred = knn_i.test(test)
    return knn_i_pred, knn_i

knn_i_pred, knn_i = knn_item(trainset, testset)
print('RMSE value(Item-based Model, test-set): ', accuracy.rmse(knn_i_pred))
knn_i_rmse = accuracy.rmse(knn_i_pred)
```

KNN User based

```python
def knn_user(train, test):
    knn_u = KNNWithMeans(k=10, sim_options={'name': 'pearson_baseline', 'user_based': True})
    knn_u.fit(train)
    knn_u_pred = knn_u.test(test)
    return knn_u_pred, knn_u

knn_u_pred, knn_u = knn_user(trainset, testset)
print('RMSE value(User-based Model, test-set): ', accuracy.rmse(knn_u_pred))
knn_u_rmse = accuracy.rmse(knn_u_pred)
```

```python
from collections import defaultdict
```

```python
def get_top_n(predictions, n=5):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # sorting predictions for each user and retrieving the n highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
top_5 = get_top_n(knn_i_pred,5)
print('Top 5 recommendations for all test users are: \n')
for key, value in top_5.items():
    print(key,'-> ', value, '\n')
```
